```python
#load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# set up figure size and dpi for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

## Legends

```python
nums = np.arange(0,10,0.1)
plt.plot(nums,np.sin(nums),label=r'$sin(x)$')
plt.plot(nums,np.cos(nums),label=r'$cos(x)$')
plt.plot(nums,np.tan(nums),label=r'$tan(x)$')
plt.legend()
plt.ylim(-2,2)
```

```python
#Location
plt.plot(nums,np.sin(nums),label=r'$sin(x)$')
plt.plot(nums,np.cos(nums),label=r'$cos(x)$')
plt.plot(nums,np.tan(nums),label=r'$tan(x)$')
# plt.legend(loc='upper right')
# plt.legend(loc='upper left')
# plt.legend(loc='center right')
plt.legend(loc='best')
plt.ylim(-2,2)
```

```python
# ncols -> to set the no. of columns for legend()
plt.plot(nums,np.sin(nums),label=r'$sin(x)$')
plt.plot(nums,np.cos(nums),label=r'$cos(x)$')
plt.plot(nums,np.tan(nums),label=r'$tan(x)$')
plt.plot(nums,np.sinh(nums),label=r'$sinh(x)$')
plt.plot(nums,np.cosh(nums),label=r'$cosh(x)$')
plt.plot(nums,np.tanh(nums),label=r'$tanh(x)$')
plt.legend()
# plt.legend(ncol=2,loc='upper right')
plt.gcf().set_size_inches(6,2)
plt.ylim(-2,2)
```

```python
#Title -> for legend title
plt.plot(nums,np.sin(nums),label=r'$sin(x)$')
plt.plot(nums,np.cos(nums),label=r'$cos(x)$')
plt.plot(nums,np.tan(nums),label=r'$tan(x)$')
plt.legend(loc='upper left',title='Trig Functions')
plt.ylim(-2,2)
```

```python
#Appearance (frameon, fancybox, shadow, fontsize)
plt.plot(nums,np.sin(nums),label=r'$sin(x)$')
plt.plot(nums,np.cos(nums),label=r'$cos(x)$')
plt.plot(nums,np.tan(nums),label=r'$tan(x)$')
plt.legend(loc ='upper right',frameon=True,
           fancybox=False,title='Trig Function',framealpha=1,edgecolor='k',
           shadow=True
          )
plt.ylim(-2,2)
```

## Colorbars

```python
#Generate a smoothed, gaussian random field
from scipy.ndimage.filters import gaussian_filter
x = np.random.normal(size=(512,512))
random = gaussian_filter(x,sigma=10)
```

```python
#orientation
plt.imshow(random,cmap='viridis')
plt.colorbar(orientation='horizontal');
```

```python
# Label
plt.imshow(random)
plt.colorbar(orientation='horizontal',label='Gaussian Random Field')
```

```python
#Alternate method for cbar title
plt.imshow([[55,2,105],[5,44,78],[92,6,234]])
cbar = plt.colorbar(orientation='horizontal')
cbar.set_label('Gaussian Random Distribution')
```
