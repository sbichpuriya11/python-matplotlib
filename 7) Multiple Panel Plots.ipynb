```python
#load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# set up figure size and dpi for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

## Subplot

```python
nums = np.arange(0,10,0.1)
plt.plot(nums,np.sin(nums));
plt.plot(nums,np.cos(nums));
```

```python
#Subplot
plt.subplot(2,2,1) #(row number,col number, plot number within that subplot set)
plt.plot(nums,np.sin(nums))
plt.subplot(2,2,2)
plt.plot(nums,np.cos(nums))
plt.subplot(2,2,3)
plt.plot(np.sin(nums),nums)
plt.subplot(2,2,4)
plt.plot(np.cos(nums),nums)
plt.tight_layout()
```

```python
# consider 3,1 as 3*1 = 3 elements i.e 1,2,3 plot numbers
plt.subplot(3,1,1)
plt.bar(nums,height=nums**1)
plt.margins(0.01,0.1)

plt.subplot(3,1,2)
plt.bar(nums,height=nums**2)
plt.margins(0.01,0.1)

plt.subplot(3,1,3)
plt.bar(nums,height=nums**3)
plt.margins(0.01,0.1)

plt.tight_layout()
```

```python
# plt.subplot(2,1,1)
plt.subplot(211)
plt.plot(nums,np.sin(nums))
# plt.subplot(2,1,2)
plt.subplot(212)
plt.plot(nums,np.cos(nums))
plt.tight_layout()
```

```python
# Subplot labels (The 'current axis' state)
plt.subplot(211)
plt.plot(nums,np.sin(nums))
plt.ylabel(r'$sin(x)$')

plt.subplot(212)
plt.plot(nums,np.cos(nums),color='orange')
plt.ylabel(r'$cos(x)$')


plt.xlabel('X')
plt.tight_layout()
```

```python
n_random = np.arange(1,11)


plt.subplot(211)
plt.plot(n_random,n_random**2,marker='^',linestyle='--')
plt.ylabel(r'$x^2$')
plt.xlim(2,8)

plt.subplot(212)
plt.plot(n_random,n_random**4,color='red',marker='o',linestyle='-.')
plt.xlim(2,8)
plt.ylabel(r'$x^4$')


plt.xlabel('X')
plt.tight_layout()
```

## Subplots

```python
# Subplots takes the dimensions as argument
# fig,axes = plt.subplots(1,2)
fig,axes = plt.subplots(3,2)
print(axes.shape)
plt.clf()
```

```python
#Looping through the axis array
xval = np.arange(10,45,2)
fig,axes = plt.subplots(2,2)
i=1
for ax in axes.flatten():
    ax.plot(xval,xval**i)
    ax.set_xticks([10,22,34,46])
    i +=0.5
plt.tight_layout()
```
