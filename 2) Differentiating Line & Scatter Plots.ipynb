## plot()

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
#Set up figure size and DPI for screen size
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

```python
#Basic line plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals))
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#different line styles
idx = 0 
for marker in ('-','--',':','-.'):
    plt.plot(xvals, np.sin(xvals)+idx,linestyle=marker)
    idx +=1
```

```python
#Markers
plt.plot(xvals,np.sin(xvals),'k.--')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#Marker frequency (markevery parameter)
xvals2 = np.arange(0,10,0.01)
# plt.plot(xvals2,np.sin(xvals2),marker='.',linestyle='--',color='k')
plt.plot(xvals2,np.sin(xvals2),marker='.',linestyle='--',markevery=20,color='k')
```

```python
#Marker Size
# plt.plot(xvals,np.sin(xvals),marker='o', color='orange',markersize=20)
plt.plot(xvals,np.sin(xvals),marker='o', color='orange',markersize=4)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
idx = 0
for marker in (".",",","o","v","^","<",">","1","2","3","4","8","s","p","P","*","h","H","+","x","X","D","d","|","_"):
#     plt.plot(idx,0,marker=marker,markerfacecolor='red',markeredgecolor='red')
    plt.plot(idx,0,marker=marker)
    idx = idx+1
plt.xlim(-1,idx+0.5)
```

```python
#latex marker (customized markers)
# plt.plot(xvals,np.sin(xvals),linestyle='--',color='goldenrod',marker=r'$0$',markevery=5)
plt.plot(xvals,np.sin(xvals),linestyle='--',color='goldenrod',marker=r'$$',markevery=5,markeredgecolor='black',markeredgewidth=0.6)
# plt.plot(xvals,np.sin(xvals),linestyle='--',color='goldenrod',marker=r'$sin(x)$',markevery=5,markeredgecolor='black',markersize=25)
# plt.plot(xvals,np.sin(xvals),linestyle='--',color='goldenrod',marker=r'$sigma = Sigma$',markevery=5,markeredgecolor='black',markersize=25)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#line style
# plt.plot(xvals,np.sin(xvals),linewidth=10,color='dodgerblue')
plt.plot(xvals,np.sin(xvals),linewidth=2,color='dodgerblue')
plt.xlabel(r'x')
plt.ylabel(r'sin(x)')
```

```python
#line style with zorder (from the different line plots which has the biggest zorder values will be in the front of the other) 
plt.plot(xvals,np.sin(xvals),linewidth=6,color='goldenrod',zorder=20)
plt.plot(xvals+1,np.sin(xvals),linewidth=6,color='navy',zorder = 2)
plt.xlabel(r'x')
plt.ylabel(r'sin(x)')
```

```python
#transparency in line (0<alphaValue<1)
plt.plot(xvals,np.sin(xvals),linewidth=6,color='red',alpha=0.2)
plt.plot(xvals+1,np.sin(xvals),linewidth=6,color='blue',alpha=0.5)
plt.xlabel(r'x')
plt.ylabel(r'sin(x)')
```

```python
#array order matters
shuffled_xvals = np.random.permutation(xvals)
#permutation will change the positions of xvals elements randomly
shuffled_xvals
```

```python
plt.plot(shuffled_xvals,np.sin(shuffled_xvals),linestyle='--',linewidth=0.9,color='black')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.plot(shuffled_xvals,np.sin(shuffled_xvals),marker='.',linestyle='none',markerfacecolor='black',markeredgecolor='black')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

## Scatter()

```python
#basic scatter plot
xvals = np.arange(0,10,0.1)
plt.scatter(xvals,np.sin(xvals),color='blue',edgecolor='black')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
#Colors
xvals = np.arange(0,10,0.1)
# plt.scatter(xvals,np.sin(xvals),c='red', edgecolor='black')
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals), edgecolor='black')
plt.colorbar()
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
# colour map
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals), cmap='inferno',edgecolor='black')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals), cmap='inferno',edgecolor='none')
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals), cmap='inferno',edgecolor='none',marker ='o',s=4)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```

```python
# sizes
# plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals), cmap='inferno',edgecolor='none',s=xvals**3)
plt.scatter(xvals,np.sin(xvals),c=np.cos(xvals), cmap='inferno',edgecolor='black',s=xvals**3)
plt.xlabel(r'$x$')
plt.ylabel(r'$sin(x)$')
```
