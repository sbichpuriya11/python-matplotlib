```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#Set up figure size and DPI for screen size
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

```python
#Basic line plot
xvals = np.arange(0,10,0.1)
plt.plot(xvals,np.sin(xvals))
```

```python
#Axis Labels
plt.plot(xvals,np.sin(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
#Multiple plots
plt.plot(xvals,np.sin(xvals))
plt.plot(xvals,np.cos(xvals))
```

```python
# Legends & Title
plt.plot(xvals,np.sin(xvals),label=r'$y = \sin(x)$')
plt.plot(xvals,np.cos(xvals),label=r'$y = \cos(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='upper right')
```

```python
#line color
plt.plot(xvals,np.sin(xvals),label=r'$y = \sin(x)$',color='red')
plt.plot(xvals,np.cos(xvals),label=r'$y = \cos(x)$',color='blue')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#line color short forms
plt.plot(xvals,np.sin(xvals),label=r'$y = \sin(x)$',color='k')
plt.plot(xvals,np.cos(xvals),label=r'$y = \cos(x)$',color='g')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#line Style & marker
plt.plot(xvals,np.sin(xvals),label=r'$y = \sin(x)$',color='red',marker = 'o')
plt.plot(xvals,np.cos(xvals),label=r'$y = \cos(x)$',color='blue')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#line style
plt.plot(xvals,np.sin(xvals),label=r'$y = \sin(x)$',color='red',linestyle='-.',)
plt.plot(xvals,np.cos(xvals),label=r'$y = \cos(x)$',color='blue',linestyle='--')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#line style with marker
plt.plot(xvals,np.sin(xvals),label=r'$y = \sin(x)$',color='red',linestyle='-.',)
plt.plot(xvals,np.cos(xvals),label=r'$y = \cos(x)$',color='blue',linestyle='none',marker='o')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best') 
```

```python
#short format code
plt.plot(xvals,np.sin(xvals),'r+--',label=r'$y = \sin(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#axis limits
plt.plot(xvals,np.sin(xvals),'r--',label=r'$y = \sin(x)$')
plt.plot(xvals,np.cos(xvals),'bo',label=r'$y = \cos(x)$')
plt.ylim(0,1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#axis limits
plt.plot(xvals,np.sin(xvals),'r--',label=r'$y = \sin(x)$')
plt.plot(xvals,np.cos(xvals),'bo',label=r'$y = \cos(x)$')
plt.ylim(0,1)
plt.xlim(0,1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(loc='best')
```

```python
#Generate a smoothed, gaussian random fields
from scipy.ndimage.filters import gaussian_filter
n_random = gaussian_filter(np.random.normal(size=(100,100)),sigma = 10)
print(n_random)
```

```python
#show image
# plt.imshow(n_random,cmap='viridis')
# plt.imshow(n_random,cmap='inferno')
plt.imshow(n_random,cmap='magma')
plt.colorbar()
```
