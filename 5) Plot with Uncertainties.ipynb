```python
#load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# set up figure size and dpi for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

## Uncertainties on Line & Scatter Plots

```python
nums = np.arange(0,10,0.5)
xerrors = np.random.uniform(size=20)
yerrors = np.random.uniform(size=20)
```

```python
plt.plot(nums,np.sin(nums))
```

```python
#Basic error plot
plt.errorbar(nums,np.sin(nums),xerr = xerrors,yerr = yerrors)
```

```python
#Errorbar colors
# plt.errorbar(nums,np.sin(nums),xerr = xerrors,yerr = yerrors,ecolor='red')
plt.errorbar(nums,np.sin(nums),xerr = xerrors,yerr = yerrors,ecolor='red',color='black')
```

```python
#Error Line width
# plt.errorbar(nums,np.sin(nums),xerr = xerrors,yerr = yerrors,ecolor='red',elinewidth=4)
plt.errorbar(nums,np.sin(nums),xerr = xerrors,yerr = yerrors,ecolor='red',elinewidth=0.4) 
```

```python
#Error bar Cap size (that horizontal size of caps)
plt.errorbar(nums,np.sin(nums),xerr = xerrors,yerr = yerrors,ecolor='red',elinewidth=0.4,capsize=2)
```

## Uncertainties on Bar Plots

```python
plt.bar(nums,np.sin(nums)**2,width=0.2,xerr = xerrors,yerr = yerrors)
```
