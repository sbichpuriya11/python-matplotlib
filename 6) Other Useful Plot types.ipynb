```python
#load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# set up figure size and dpi for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

## Fill & Fill Between

```python
#fill
xvals = np.arange(0,10,0.1)
plt.fill(xvals,np.sin(xvals),edgecolor='k')
```

```python
#Fill Color
plt.fill(xvals,np.sin(xvals),color='orange',edgecolor='k')
```

```python
plt.fill(xvals,np.sin(xvals),color='orange',edgecolor='k',hatch='x')
```

```python
#Fill Between
plt.fill_between(xvals,xvals+np.sin(xvals),np.sin(xvals)-xvals,edgecolor='k')
```

## Hexbin & Hist2D

```python
random_x = np.random.normal(size=int(1e5))
random_y = np.random.normal(size=int(1e5))
```

```python
plt.plot(random_x,random_y,linestyle='none',marker=',',color='black');
```

```python
#Hexbin
plt.hexbin(random_x,random_y)
```

```python
#bins (decides the nummber of groups of element)
plt.hexbin(random_x,random_y,bins=4)
```

```python
plt.hexbin(random_x,random_y,bins=10)
```

```python
#Mincnt
plt.hexbin(random_x,random_y,bins=5,mincnt= 5)
```

```python
#Hist2D
plt.hist2d(random_x,random_y);
```

```python
plt.hist2d(random_x,random_y,bins=40);
```
