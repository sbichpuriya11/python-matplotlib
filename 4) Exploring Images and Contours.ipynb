## imshow()

```python
#load numpy for math/array operations
# and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# set up figure size and dpi for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

```python
#Generate a smoothed, gaussian random field
from scipy.ndimage.filters import gaussian_filter
n_random = gaussian_filter(np.random.normal(size=(512,512)),sigma=10)
```

```python
#Basic imshow
plt.imshow(n_random)
plt.colorbar();
```

```python
# Extent & LRBT (extent = [left,right,top,bottom])
plt.imshow(n_random,extent=[1,5,5,1]); 
plt.colorbar()
```

```python
#Max/Min value for color from colorbar, these values are exclusive
plt.imshow(n_random,vmin=-0.075,vmax=0.05)
plt.colorbar()
```

```python
#Aspect ratio (vertical vs horizontal axis)
# aspect <1 is vertial height
# aspect >1 is horizontal width
#aspect =1 gives squared image
plt.imshow(n_random,aspect=0.3)
plt.colorbar()
```

```python
#Interpolation
# vales for interpolation
#antialiased, none, nearest, bilinear, bicubic, spline16, spline36, hanning, 
# hamming, hermite, kaiser, quadric, catrom, gaussian, bessel, mitchell, sinc, lanczos, blackman
small = np.arange(0,4,1).reshape(2,2)
print(small)
plt.imshow(small,interpolation='quadric')
plt.imshow(small,interpolation='gaussian')
```

```python
# It disables the interpolation and shift the axis ticks to center
plt.matshow(small)
```

## Contours

```python
#Basic contours
s_random = gaussian_filter(n_random,sigma = 100)
print(s_random)
```

```python
plt.imshow(s_random)
plt.colorbar()
```

```python
plt.contour(s_random)
plt.colorbar();
```

```python
#Contours Level(must be in increasing order)
plt.contour(s_random,levels=[-0.0045,-0.0030 ,-0.0015,0,0.003])
plt.colorbar();
```

```python
li = []
for i in range(1,100):
    temp = i/1000
    li.append(temp)
# plt.contour(s_random,levels=[0.0045,0.0046,0.0047])
plt.contour(s_random,levels=li)
plt.colorbar();
```

```python
#Nummber of contours
# plt.contour(s_random,30)
plt.contour(s_random,10)
```

```python
#contours Colors
plt.contour(s_random,10,colors=['b','y'])
plt.contour(s_random,15,colors=['b','y','orange','k'])
```

```python
#Filled contours
plt.contourf(s_random,10);
```

```python
plt.contourf(s_random,10,colors=['b','g','b','y','r'])
plt.colorbar()
```

```python
#Hatches
plt.contourf(s_random,10,colors='none',hatches=['o','y'])
```

```python
#Line Style and widths
plt.contour(s_random,linewidths=[1,5],linestyles=[':','-','--']);
```

```python
plt.contour(s_random,linewidths=[1,3],colors=['blue','black','green'],linestyles=['-.','-','--']);
```

```python
# Contours Labels (to throw the labels for the contours)
contours = plt.contour(s_random,5)
plt.clabel(contours)
```

```python
contours = plt.contour(s_random,colors='k',linestyles=['-.','--'])
plt.clabel(contours)
```

```python
plt.contour(s_random,levels=[0],colors=['k'])
plt.imshow(s_random)
plt.colorbar();
```
